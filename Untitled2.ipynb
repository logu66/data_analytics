```python
import requests

from bs4 import BeautifulSoup
```

```python
url='https://www.flipkart.com/search?q=fogg+perfume&sid=g9b%2C0yh%2Cvp1%2C0kb&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&as-pos=1&as-type=RECENT&suggestionId=fogg+perfume%7CDeodorant+Spray&requestId=c7188b46-7d79-4267-b715-786ce521a25e&as-searchtext=fog'
```

```python
response=requests.get(url)

response
```
